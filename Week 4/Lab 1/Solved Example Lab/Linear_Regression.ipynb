```python
cars = pd.read_csv('carprices.csv')
```

```python
cars
```

```python
cars.columns = ['Car_Model', 'Milage', 'Sell_Price', 'Age']
cars
```

##One‑hot encode the Car_Model column to convert into numerical data for traininng the model.

The model can only be trained on numerical data

```python
dummies = pd.get_dummies(cars.Car_Model)
dummies
```

```python
dum_cars = pd.concat([cars,dummies], axis = 'columns')
dum_cars
```

```python
dum_cars.rename(columns={'Mercedez Benz C class': 'Mercedes_C_class'}, inplace=True)
```

```python
dum_cars
```

##Drop the contextual column

```python
dum_cars = dum_cars.drop('Car_Model', axis='columns')
```

```python
l_model = linear_model.LinearRegression()
```

```python
y = cars.Sell_Price
print(y)
X = dum_cars.drop('Sell_Price', axis = 'columns')
X
```

#Now train this data on linear regression and predict price for 45000 milage,4 years old Mercedes C class and 86000 milage, 7 years old Audi A5

```python
#Code here
import pandas as pd

l_model.fit(X, y)

# The columns in X are ordered as: ['Milage', 'Age', 'Audi A5', 'BMW X5', 'Mercedes_C_class']
# We wrap the prediction inputs in a pandas DataFrame with matching column names to avoid model warnings.

# 1. Predict price for 45,000 mileage, 4 years old, Mercedes C class
# Values: Milage=45000, Age=4, Audi A5=0, BMW X5=0, Mercedes_C_class=1
mercedes_data = pd.DataFrame([[45000, 4, 0, 0, 1]], columns=X.columns)
mercedes_pred = l_model.predict(mercedes_data)

# 2. Predict price for 86,000 mileage, 7 years old, Audi A5
# Values: Milage=86000, Age=7, Audi A5=1, BMW X5=0, Mercedes_C_class=0
audi_data = pd.DataFrame([[86000, 7, 1, 0, 0]], columns=X.columns)
audi_pred = l_model.predict(audi_data)

print(f"Predicted price for 45000 mileage, 4 years old Mercedes C class: ${mercedes_pred[0]:,.2f}")
print(f"Predicted price for 86000 mileage, 7 years old Audi A5: ${audi_pred[0]:,.2f}")
```
