#Decision Tree Classifier Lab


#Load essential libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
```

#Import the dataset

```python
df = pd.read_csv("titanic.csv")
df
```

##Drop meaningless information

```python
df = df.drop(columns = ['PassengerId','Name','SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns')
df
```

##Data Preprocessing

```python
import math
mean_age = df.Age.mean()
mean_age = math.floor(mean_age)
print(mean_age)
```

```python
df['Age'] = df.Age.fillna(mean_age)
df
```

##Label Encode the non numeric features

```python
LE_sex = LabelEncoder()
df['Sex'] = LE_sex.fit_transform(df['Sex'])

df
```

## Define Feature Matrix (X) and Target Vector (y)


```python
y = df['Survived']
X = df.drop('Survived', axis = 'columns')
X
```

## Train-Test Split and Model Initialization


```python
model = tree.DecisionTreeClassifier()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)
```

##Train the model

```python
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

##Predict results

```python
model.predict(X_test)
```

##Try decision tree classifer on the same target column but this time use more feature columns than before.

```python
df2 = pd.read_csv("titanic.csv")

df2 = df2.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], axis='columns')

import math
mean_age = math.floor(df2['Age'].mean())
df2['Age'] = df2['Age'].fillna(mean_age)

mode_embarked = df2['Embarked'].mode()[0]
df2['Embarked'] = df2['Embarked'].fillna(mode_embarked)

from sklearn.preprocessing import LabelEncoder
LE_sex = LabelEncoder()
LE_embarked = LabelEncoder()

df2['Sex'] = LE_sex.fit_transform(df2['Sex'])
df2['Embarked'] = LE_embarked.fit_transform(df2['Embarked'])

y2 = df2['Survived']
X2 = df2.drop('Survived', axis='columns')

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.4, random_state=42)
model2 = tree.DecisionTreeClassifier(random_state=42)

model2.fit(X_train2, y_train2)

new_score = model2.score(X_test2, y_test2)
print(f"Model Accuracy with added features: {new_score:.4f}")
```
